```python
import matplotlib.pyplot as plt
import numpy as np

# データの準備
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.randn(100)

# サブプロットの作成 (2行3列)
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 各サブプロットへのプロット
# 折れ線グラフ
axes[0, 0].plot(x, y1)
axes[0, 0]
# sin関数
axes[0, 1].plot(x, np.sin(1* np.pi * x))
axes[0, 1]
# ヒストグラム
axes[0, 2].hist(y2, bins=20)
axes[0, 2]
# 散布図
axes[1, 0].scatter(x, y2)
axes[1, 0]

# グラフの表示
plt.show()
```
